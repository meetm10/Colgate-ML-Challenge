```python
import pandas
import math
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import cross_validate
from sklearn import datasets, linear_model
```

```python
file = pandas.read_csv("hack_ru(1).csv")
```

```python
final_list_ingredient = []
```

```python
file["ingredients"] = file["ingredients"].fillna("na")
```

```python
file = file.fillna(0.0)
```

```python
for i in range(len(file)):
    a =[x.replace(" ","").lower() for x in file["ingredients"][i].split(",")]
    final_list_ingredient.extend(a)
```

```python
final_list_ingredient = list(dict.fromkeys(final_list_ingredient))
```

```python
final_list_ingredient
```

```python
final_ingredient_feature =[]
for i in range(len(file)):
    print (i)
    row_feature = []
    row = file["ingredients"][i].lower()
    for j in range(len(final_list_ingredient)):
        
        if final_list_ingredient[j] in row:
            row_feature.append(1)
        else:
            row_feature.append(0)        
    final_ingredient_feature.append(row_feature)
```

```python
ingredient_array = np.asarray(final_ingredient_feature)
```

```python
ingredient_array.shape
```

```python
temp = file[['company','country']]
```

```python
le = preprocessing.LabelEncoder()
X_2 = temp.apply(le.fit_transform)
```

```python
enc = preprocessing.OneHotEncoder(sparse=True)
```

```python
enc.fit(X_2)
```

```python
onehotlabels = enc.transform(X_2)
```

```python
a = [float(i) for i in list(file['total_pack_size_ml_g'])]
```

```python
b = max(a)
```

```python
a = [i/b for i in a ]
```

```python
dense_one_hot = np.asarray(onehotlabels.todense())
```

```python
dense_one_hot.shape, ingredient_array.shape
```

```python
temp = np.concatenate((ingredient_array,dense_one_hot),axis=1)
```

```python
np.asarray(a).shape,temp.shape
```

```python
data = np.concatenate((temp,np.asarray(a).reshape(-1,1)),axis=1)
```

```python
a = [float(i) for i in list(file['price_per_100g_ml_dollars'])]
b = max(a)
target = [i/b for i in a ]
```

```python
target = np.asarray(target).reshape(-1,1)
```

```python
lasso = linear_model.Lasso()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     data, target, test_size=0.2, random_state=42)
```

```python
clf = linear_model.Lasso(alpha=0.1)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
np.sum((y_test-y_pred)**2)/len(y_test)
```
